```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
data_root = 'D:\ml\examples\datasets'
hap = pd.read_csv(data_root + '\income.data.csv')
x = hap[['income']]
y = hap[['happiness']]
```

```python
hap.plot(kind = 'scatter', grid  = True, 
        x = 'income', y = 'happiness')
plt.show()
```

Linear Regression

use .values attributes of the training sets to avoid any "UserWarning: X does not have valid feature names, but LinearRegression was fitted with feature names" error. you still get the output either way, but this is a more elegant compatible way.

```python
model1 = LinearRegression()
model1.fit(x.values,y.values)
```

```python
x_new = [[4.8]]
print(model1.predict(x_new))
```

KNN

use .values attributes of the training sets to avoid any "UserWarning: X does not have valid feature names, but LinearRegression was fitted with feature names" error. you still get the output either way, but this is a more elegant compatible way.

```python
model2 = KNeighborsRegressor(n_neighbors = 3)
model2.fit(x.values,y.values)
print(model2.predict(x_new))
```
